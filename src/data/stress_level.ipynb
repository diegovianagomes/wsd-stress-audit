```python
import pandas as pd
import os
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
stress_level = pd.read_csv('Stress_Level_v2.csv')
stress_level 
```

```python
import seaborn as sns
```

```python
stress_level.reset_index(inplace=True)
stress_melted = stress_level.melt(
    id_vars=['index'],
    value_vars=['Baseline', 'TMCT', 'First Rest', 'Real Opinion', 'Opposite Opinion', 'Second Rest', 'Subtract'],
    var_name='Protocol',
    value_name='Stress_Level'
)
stress_melted = stress_melted.rename(columns={'index': 'Subject'})
stress_melted
```

```python
plt.figure(figsize=(12, 8))
sns.lineplot(data=stress_melted, x='Protocol', y='Stress_Level', hue='Subject', marker='o')
plt.xticks(rotation=45)
plt.title('Níveis de Stress por Participante e Protocolo')
plt.xlabel('Protocolo')
plt.ylabel('Nível de Stress')
plt.legend(title='Participante', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
